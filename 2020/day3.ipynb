```python
# lines = [
#     '..##.......',
#     '#...#...#..',
#     '.#....#..#.',
#     '..#.#...#.#',
#     '.#...##..#.',
#     '..#.##.....',
#     '.#.#.#....#',
#     '.#........#',
#     '#.##...#...',
#     '#...##....#',
#     '.#..#...#.#'
# ]
#
with open('data/day3') as f:
    lines = list(f.read().splitlines())

lines
```

```python
# Get string result at the given index
def pos_at(pattern, index):
    pattern_position = index % len(pattern)
    return pattern[pattern_position]

def map_coordinates(x_y):
    (x, y) = x_y

    if y >= len(lines):
        return 0

    character = pos_at(lines[y], x)

    return 1 if character == '#' else 0


coordinates = list(map(lambda x: (x * 3, x), range(1, len(lines))))

result = sum(map(map_coordinates, coordinates))

result
```

```python
import functools

slopes = [
    (1, 1),
    (3, 1),
    (5, 1),
    (7, 1),
    (1, 2)
]

def map_slope(slope):
    (deltaX, deltaY) = slope

    return list(map(lambda x: (x * deltaX, x * deltaY), range(1, len(lines))))

coordinates = map(map_slope, slopes)

total_result = []

for current_coordinates in coordinates:
    result = sum(map(map_coordinates, current_coordinates))
    total_result.append(result)


import operator
print(functools.reduce(operator.mul, total_result, 1))
```
