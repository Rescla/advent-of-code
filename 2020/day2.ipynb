```python
lines = []

with open('data/day2') as f:
    lines = list(map(lambda x: x.split(), f.read().splitlines()))

lines
```

```python
count = 0

for line in lines:
    length_policy = line[0].split('-')

    at_least = int( length_policy[0])
    at_most = int( length_policy[1])

    character = line[1][0]

    character_count = line[2].count(character)

    print(character_count)

    if at_least <= character_count <= at_most:
        count += 1

count
```

```python
# lines = [
#     '1-3 a: abcde',
#     '1-3 b: cdefg',
#     '2-9 c: ccccccccc'
# ]
#
# lines = list(map(lambda x: x.split(), lines))

count = 0

for line in lines:
    length_policy = line[0].split('-')
    password = line[2]

    should_match_i = int( length_policy[0]) - 1
    should_not_match_i = int( length_policy[1]) - 1

    if should_match_i >= len(password) or should_not_match_i >= len(password):
        print('not valid', line)
        continue

    should_match = password[should_match_i]
    should_not_match = password[should_not_match_i]

    character = line[1][0]

    if bool(character == should_match) != bool(character == should_not_match):
        print('matched', line)
        count += 1
    else:
        print('not matched', line)

count
```
