```python
def has_required_fields(passport):
    fields_in_passport = list(map(lambda item: item[0], passport))
    required_fields = ['byr', 'iyr', 'eyr', 'hgt', 'hcl', 'ecl', 'pid']

    return all(field in fields_in_passport for field in required_fields)

valid_passports = 0

for current_passport in lines:
    if has_required_fields(current_passport):
        valid_passports += 1

print('valid passports: ', valid_passports)
```

```python
def get_field(passport, field):
    return next(filter(lambda entry: entry[0] == field, passport))[1]

def is_height_valid(passport):
    height = get_field(passport, 'hgt')

    if not re.match('\d+(cm|in)', height):
        return False

    height_number = int(re.search('(\d)+', height).group(0))

    if re.match('\d+cm', height):
        return 150 <= height_number <= 193

    return 59 <= height_number <= 76


def is_valid(passport):
    if not has_required_fields(passport):
        return False

    valid = True

    byr = get_field(passport, 'byr')
    valid = valid and byr.isnumeric() and (1920 <= int(byr) <= 2002)

    iyr = get_field(passport, 'iyr')
    valid = valid and iyr.isnumeric() and (2010 <= int(iyr) <= 2020)

    eyr = get_field(passport, 'eyr')
    valid = valid and eyr.isnumeric() and (2020 <= int(eyr) <= 2030)

    valid = valid and is_height_valid(passport)

    hcl = get_field(passport, 'hcl')
    valid = valid and re.match('^#[0-9a-f]{6}$', hcl)

    ecl = get_field(passport, 'ecl')
    valid = valid and re.match('^(amb|blu|brn|gry|grn|hzl|oth)$', ecl)

    pid = get_field(passport, 'pid')
    valid = valid and re.match('^\d{9}$', pid)

    return valid


valid_passports = 0

for current_passport in lines:
    if is_valid(current_passport):
        valid_passports += 1

print('valid: ', valid_passports)
```
